```python
a_movies = pd.read_csv("../scraper/data/movies_metadata.csv")
```

```python
from itertools import groupby

base_dir = "C:\\Users\\barto\Desktop\\PosterRecognition\\scraper\\data\\images"

num = {"tt{:0>7}".format(p_dir): len(os.listdir(base_dir + f"\\{p_dir}")) for p_dir in os.listdir(base_dir)}

num
```

```python
grouped = {}

for key, group in groupby(sorted(num.items(), key=lambda y: y[1]), lambda x: x[1]):
    grouped[key] = [item[0] for item in group]
```

```python
pd.DataFrame(list(num.items()), columns=['imdb_id', 'poster_count']).groupby('poster_count').count()
```

```python
a_movies['poster_count'] = a_movies['imdb_id'].map(num)
a_movies = a_movies.dropna(subset=['poster_count'])
a_movies.head()
```

```python
a_movies.to_csv("../scraper/data/movies_with_posters.csv")
```

```python
#dataframe containing movies with avaliable posters along with rich descriptions from different table.
posters_descriptions = pd.merge(m_descr, a_movies, how='inner', on='imdb_id')
columns_to_be_deleted = (set(posters_descriptions.columns) & set(m_descr)) - set(['plot_synopsis', 'tags'])

posters_descriptions.drop(axis=1, columns=columns_to_be_deleted, inplace=True)
```

```python
posters_descriptions['num_description_tokens'] = posters_descriptions['plot_synopsis'].apply(lambda string: len(string.split(" ")))
posters_descriptions['num_description_tokens']
```

```python
posters_descriptions['title'] = posters_descriptions['title_y']
posters_descriptions.drop(axis=1, columns=['title_x', 'title_y'], inplace=True)
```
